# Q-2. Write an SQL query to fetch duplicate records having matching data in some fields of a table.

```python
data_2 = pd.read_sql("SELECT *  FROM employees.worker  GROUP BY First_Name HAVING COUNT(*) > 1;",mydb)
print(data_2)
```
