# Q-19. Write an SQL query to determine the 5th highest salary without using TOP or limit method.

```python
data_1 = pd.read_sql("SELECT * FROM ( SELECT ROW_NUMBER() OVER (ORDER BY Salary DESC) AS rownumber,Salary FROM employees.worker ) AS foo WHERE rownumber = 5",mydb)
data_1
```

# Q-20. Write an SQL query to fetch the list of employees with the same salary.

```python
data_2 = pd.read_sql("SELECT First_Name,Salary FROM employees.worker WHERE Salary IN ( SELECT Salary FROM employees.worker GROUP BY Salary HAVING COUNT(*) > 1)",mydb)
data_2
```
