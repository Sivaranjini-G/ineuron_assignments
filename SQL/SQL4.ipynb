```python
import pandas as pd
import mysql.connector as conn

mydb = conn.connect(host = 'localhost', user = 'root', password="root")
cursor = mydb.cursor()
```

# Q-1. Write an SQL query to print the FIRST_NAME from Worker table after replacing ‘a’ with ‘A’.

```python
data_1 = pd.read_sql("SELECT REPLACE(First_Name, 'a', 'A') FROM employees.worker", mydb)
data_1
```

# Q-2. Write an SQL query to print all Worker details from the Worker table order by FIRST_NAME Ascending and DEPARTMENT Descending.

```python
data_2 = pd.read_sql("select * from employees.Worker ORDER BY First_Name ASC , Department Desc ", mydb)
data_2
```

# Q-3. Write an SQL query to fetch the names of workers who earn the highest salary.

```python
data_3 = pd.read_sql("select first_name, last_name , max(salary) from employees.worker",mydb)
data_3
```
