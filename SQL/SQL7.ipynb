```python
data_1 = pd.read_sql("SELECT * FROM ( SELECT *,ROW_NUMBER() OVER() AS rownum FROM employees.worker ) t WHERE t.rownum % 2 != 0",mydb)
data_1
```

# Q-14. Write an SQL query to clone a new table from another table.

```python
cursor.execute("CREATE TABLE employees.workerclone SELECT * FROM employees.worker")
data_2 = pd.read_sql("SELECT * FROM employees.workerclone",mydb)
data_2
```
